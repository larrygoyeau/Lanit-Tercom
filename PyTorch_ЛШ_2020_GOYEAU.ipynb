```
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```
data = pd.read_csv("dataset.csv", encoding='utf-8')
data.head()
```

```
# данные скачанного датасета
plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=data['target'], cmap='brg')
plt.title('Гавиал и крокодил', fontsize=15)
plt.xlabel('Тон окраски', fontsize=14)
plt.ylabel('Ширина мордочки', fontsize=14)
plt.show();
```

На рисунке разнвми цветами показаны 2 класса объектов. 

```
# массив с признаками (2 признака)
X = data.iloc[:,:2].values[:]  
X[:5] 
```

```
# массив с классами (0 или 1, крокодил или гавиал)
Y = data['target'].values.reshape((-1, 1))[:] 
Y[:5]
```

```
# добавляем библиотеку
import torch
```

```
num_features = X.shape[1]
```

Создаем нейрон с помощью torch.nn.Sequential - контейнер, в котором задаются методы в том порядке, в котором должны будут выполняться.

Переменная **neuron** является объектом класса **Sequential** и состоит из одного линейного слоя и сигмоиды.

```
neuron = torch.nn.Sequential(
    torch.nn.Linear(num_features, out_features=1),
    torch.nn.Sigmoid()
)
neuron
```

Теперь **neuron** можно применить к тензору. В данном случае начальные веса инициализируются случайным образом.



```
neuron(torch.autograd.Variable(torch.FloatTensor([1, 1])))
```

Протестируем созданный нейрон на нашем датасете:

```
proba_pred = neuron(torch.autograd.Variable(torch.FloatTensor(X)))
y_pred = proba_pred > 0.5
y_pred = y_pred.data.numpy().reshape(-1)

plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=y_pred, cmap='brg')
plt.title('Гавиал и крокодил', fontsize=15)
plt.xlabel('Тон окраски', fontsize=14)
plt.ylabel('Ширина мордочки', fontsize=14)
plt.show();
```

###Теперь приступим к обучению сети!

Каждый из входных признаков обернем в тензоры (FloatTensor), а тензоры в класс autograd.Variable, который позволит при вызове функции .backward() автоматически вычислять градиент тензора.

```
X = torch.autograd.Variable(torch.FloatTensor(X))
Y = torch.autograd.Variable(torch.FloatTensor(Y))
```

Код ниже - основной, с которым придется работать при выполнении второго самастоятельного задания.


```
# функция потерь (квадратичная)
loss_func = torch.nn.MSELoss(size_average=False)

# скорость обучения (шаг метода оптимизации) 
learning_rate = 1e-3  # == 0.001 

# метод оптимизации (SGD - стахостический градиентный спуск)
optimizer = torch.optim.SGD(neuron.parameters(), lr=learning_rate)

# количество итераций
num_epochs = 500

# обучение
for i in range(num_epochs):
    y_pred = neuron(X)

    # Вычисление потер
    loss = loss_func(y_pred, Y)
    print(f'Iter: {i} -> Loss: {loss.data}')

    # обнуление градиентов, иначе будут копиться градинты с предыдущей итерации
    optimizer.zero_grad()

    # вычисление градиентов
    loss.backward()

    # обновление весов
    optimizer.step()
```

Тестируем:

```
proba_pred = neuron(X)
y_pred = proba_pred > 0.5
y_pred = y_pred.data.numpy().reshape(-1)

plt.figure(figsize=(10, 8))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=y_pred, cmap='brg')
plt.title('Гавиал и крокодил', fontsize=15)
plt.xlabel('Тон окраски', fontsize=14)
plt.ylabel('Ширина мордочки', fontsize=14)
plt.show();
```

```
neuron = torch.nn.Sequential(
    torch.nn.Linear(num_features, out_features=1),
    torch.nn.Sigmoid()
)

# функция потерь (квадратичная)
loss_func = torch.nn.MSELoss(size_average=False)

# скорость обучения (шаг метода оптимизации) 
learning_rate = 1e-2  # == 0.01

# метод оптимизации (SGD - стахостический градиентный спуск)
optimizer = torch.optim.SGD(neuron.parameters(), lr=learning_rate)

# количество итераций
num_epochs = 500

loss_list=[]
# обучение
for i in range(num_epochs):
    y_pred = neuron(X)

    # Вычисление потер
    loss = loss_func(y_pred, Y)
    loss_list=loss_list+[loss]

    # обнуление градиентов, иначе будут копиться градинты с предыдущей итерации
    optimizer.zero_grad()

    # вычисление градиентов
    loss.backward()

    # обновление весов
    optimizer.step()

plt.plot(loss_list)
plt.title('Эволюция потери со скоростью обучения '+str(learning_rate))
```

#Задание 3

Теперь попробуем классифицировать объекты трех классов. 

```
import numpy as np

N = 100
D = 2
K = 3
X = np.zeros((N * K, D))
y = np.zeros(N * K, dtype='uint8')

for j in range(K):
    ix = range(N * j,N * (j + 1))
    r = np.linspace(0.0, 1, N)
    t = np.linspace(j * 4, (j + 1) * 4,N) + np.random.randn(N) * 0.2
    X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
    y[ix] = j

plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.brg)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();
```

```
X = torch.autograd.Variable(torch.FloatTensor(X))
y = torch.autograd.Variable(torch.LongTensor(y.astype(np.int64)))
```

```
print(X.data.shape, y.data.shape)
```

```
ndim_in = 2 # размерность входа (то есть количество признаков)
ndim_out = 3 # размерность выходного слоя (то есть количество классов)
num_epochs = 1000
learning_rate = 1e-2

neuron = torch.nn.Sequential(
    torch.nn.Linear(ndim_in, ndim_out),
)

loss_fn = torch.nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(neuron.parameters(), lr=learning_rate)

loss_list=[]

for i in range(num_epochs):
    y_pred = neuron(X)

    loss = loss_fn(y_pred, y)
    loss_list=loss_list+[loss]

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

plt.plot(loss_list)
plt.title('Эволюция потери')
```

```
# приводим к типу Numpy для отрисовки
X = X.data.numpy()
y = y.data.numpy()
```

```
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
grid_tensor = torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()])

Z = neuron(torch.autograd.Variable(grid_tensor))
Z = Z.data.numpy()
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))

plt.contourf(xx, yy, Z, cmap=plt.cm.rainbow, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.brg)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();
```
